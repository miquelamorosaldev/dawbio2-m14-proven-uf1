## 1.Explicación del contexto.
Un brote de enfermedad transmitida por los alimentos ocurre cuando dos o más personas contraen la misma enfermedad por la misma comida o bebida contaminada. Si bien la mayoría de las enfermedades transmitidas por los alimentos no son parte de un brote reconocido, los brotes brindan información importante sobre cómo se propagan los gérmenes, qué alimentos causan enfermedades y cómo prevenir infecciones.

Las agencias de salud pública en los 50 estados, el Distrito de Columbia, los territorios de EE. UU. y los Estados libremente asociados tienen la responsabilidad principal de identificar e investigar brotes y utilizan un formulario estándar para informar brotes voluntariamente a los CDC. Durante 1998–2008, la notificación se realizó a través del Sistema electrónico de notificación de brotes transmitidos por los alimentos (eFORS).

### Qué son estos datos?
Este conjunto de datos contiene información sobre los brotes de enfermedades transmitidas por alimentos informados a los CDC (https://www.cdc.gov/)en los Estados Unidos desde 1998 hasta 2015.

## 2. Explicar les columnes que usareu (no cal totes). a) Nom b) Tipus (string, categorical, data, número enter, decimal ...) c) Per a què serveix, si no queda clar amb el nom.

**Columnas**  

Identificar tipo de dato:  

str = string  

int = integer  


__Year__ (int): Año de las incidencias.  

__Month__ (str): Mes en que ocurrieron las incidencias.  

__State__ (str): Estado Americano, en que ocurrió la incidencia. (brotes que ocurren en más de un estado se enumeran como "multiestatales").  

__Location__ (str): Lugar donde se preparó el alimento.  

__Food__ (str): Alimentos que llevan el contaminante.  

__Ingredient__ (str): Ingrediente posiblemente contaminado.  

__Species__ (str): Patógeno.  

__Serotype/Genoty__ (str): Toxina o sustancia química que causó la enfermedad.  

__Status__ (str): Si la enfermedad se confirmó o quedó en sospechoso.  

__Ilinesses__ (int): Número de enfermedades registradas.  

__Hospitalizaciones__ (int): número de hospitalizaciones registradas.  

__Muertes__ (int): Número de muertes registradas.  


En muchas investigaciones de brotes, no se identifica un vínculo alimentario específico; para estos brotes, la variable vínculo alimentario está en blanco.
